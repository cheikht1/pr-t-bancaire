```python
# Dictionnaire de correspondance entre les noms anglais et français des colonnes
noms_colonnes_fr = {
    'age': 'age',
    'job': 'emploi',
    'marital': 'statut_matrimonial',
    'education': 'education',
    'default': 'défaut',
    'housing': 'logement',
    'loan': 'pret',
    'contact': 'contact',
    'month': 'mois',
    'day_of_week': 'jour_de_la_semaine',
    'duration': 'durée',
    'campaign': 'campagne',
    'pdays': 'pdays',
    'previous': 'précédent',
    'poutcome': 'résultat_précédent',
    'emp.var.rate': 'taux_variabilité_emploi',
    'cons.price.idx': 'indice_prix_consommation',
    'cons.conf.idx': 'indice_confiance_consommateurs',
    'euribor3m': 'euribor_3m',
    'nr.employed': 'nombre_employés',
    'y': 'souscrit_dépôt_à_terme'
}

# Renommer les colonnes
data.rename(columns=noms_colonnes_fr, inplace=True)

# Afficher les types de données après le renommage
print(data.dtypes)
```

```python
# Remove missing values
data.dropna(inplace=True)

# Remove duplicates
data.drop_duplicates(inplace=True)
```

```python
data.info()
```

```python
categorical_variables = ['emploi', 'statut_matrimonial', 'education', 'défaut', 'logement', 'pret', 'contact', 'mois', 'jour_de_la_semaine', 'résultat_précédent', 'souscrit_dépôt_à_terme']

for variable in categorical_variables:
    unique_values = data[variable].value_counts()
    print(f"Unique values for {variable}:")
    print(unique_values)
    print()
```

```python
import matplotlib.pyplot as plt

# Generate visualizations for categorical variables
for variable in categorical_variables:
    unique_values = data[variable].value_counts()
    unique_values.plot(kind='bar')
    plt.title(f"Count of unique values for {variable}")
    plt.xlabel(variable)
    plt.ylabel("Count")
    plt.show()
```

```python
import matplotlib.pyplot as plt

# Generate boxplots for categorical variables
for variable in categorical_variables:
    data.boxplot(column=variable)
    plt.title(f"Boxplot for {variable}")
    plt.xlabel(variable)
    plt.ylabel("Value")
    plt.show()
```
